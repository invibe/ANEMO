# to_deg

Transforms **the position of the eye in pixels** into **degrees** 

## Test the parameters

-----------------------------

**example of code:** test the parameters of the function to ``the raw data position x of the eye in pixels``.

```python
import ANEMO

PreProcessing = ANEMO.PreProcessing(dirpath='dataANEMO', sub='001')

#------------------------------------------------------------------------------
# Parameters:
#------------------------------------------------------------------------------
xname = 'eye1_x_coordinate' # The raw position x of the eye in pixels
trial = 8 # Number of the trial to be transformed
expname = 'sub-001' # Name of the particular experience to be transformed
                    # or list(PreProcessing.data.keys())[0]
events_start = 'StimulusOff' # Event marking the start of the trial
ref_time = 'time' # Name of the reference time for events_start
events_SSACC='SSACC' # Event marking the start of the saccades in trial
events_ESACC='ESACC' # Event marking the end of the saccades in trial
ref_time_sacc = 'time' # Name of the reference time for events_SSACC
                       # and events_ESACC
before_sacc = 5 # Time to delete before saccades
after_sacc = 15 # Time to delete after saccades
toxname= 'xpr_deg' # Name of the data to be saved
return_ = True # If True returns the value, else saves it in data

#------------------------------------------------------------------------------
# to apply the function:
#------------------------------------------------------------------------------
xpr_deg = PreProcessing.Trial.to_deg(xname=xname,
                                     trial=trial,
                                     expname=expname,
                                     events_start=events_start,
                                     ref_time = ref_time,
                                     events_SSACC=events_SSACC,
                                     events_ESACC=events_ESACC,
                                     ref_time_sacc=ref_time_sacc,
                                     before_sacc=before_sacc,
                                     after_sacc=after_sacc,
                                     toxname=toxname,
                                     return_=return_)

#------------------------------------------------------------------------------
# to retrieve the raw data:
#------------------------------------------------------------------------------
data_exp = PreProcessing.data[expname]
data = data_exp.Data
xpr = data[data.trial==trial][xname].values

#------------------------------------------------------------------------------
# Plot:
#------------------------------------------------------------------------------
import matplotlib.pyplot as plt


fig, axs = plt.subplots(1, 2, figsize=(10, 10/(1.681*2)))

axs[0].plot(xpr, 'k', alpha=.7)
axs[0].set_title('the position of the eye in pixels: eye1_x_coordinate')
axs[0].set_xlabel('time (ms)')
axs[0].set_ylabel('position of the eye (px)')

axs[1].plot(xpr_deg, 'k', alpha=.7)
axs[1].set_title('the position of the eye in degrees: xpr_deg')
axs[1].set_xlabel('time (ms)')
axs[1].set_ylabel('position of the eye (°)')

plt.tight_layout()
plt.show()
```

## Run it on a data set

----------------------------

**example of code:** run this function on ``the full raw data of the x position of the eye in pixels``.

```python
import ANEMO

PreProcessing = ANEMO.PreProcessing(dirpath='dataANEMO', sub='001')

#------------------------------------------------------------------------------
# Parameters:
#------------------------------------------------------------------------------
xname = 'eye1_x_coordinate' # The raw position x of the eye in pixels
events_start = 'StimulusOff' # Event marking the start of the trial
ref_time = 'time' # Name of the reference time for events_start
events_SSACC='SSACC' # Event marking the start of the saccades in trial
events_ESACC='ESACC' # Event marking the end of the saccades in trial
ref_time_sacc = 'time' # Name of the reference time for events_SSACC
                       # and events_ESACC
before_sacc = 5 # Time to delete before saccades
after_sacc = 15 # Time to delete after saccades
toxname= 'xpr_deg' # Name of the data to be saved
expnames = 'sub-001' # Name of the particular experience to be transformed
                     # or list(PreProcessing.data.keys())[0] 
trials = list(range(1, 16)) # List of the trial to be transformed
recalculate = False # Allows you to indicate if you want to force the
                    # calculation on the trials already processed

#------------------------------------------------------------------------------
# to apply the function:
#------------------------------------------------------------------------------
PreProcessing.Data.to_deg(xname=xname,
                          events_start=events_start,
                          ref_time=ref_time,
                          events_SSACC=events_SSACC,
                          events_ESACC=events_ESACC,
                          ref_time_sacc=ref_time_sacc,
                          before_sacc=before_sacc,
                          after_sacc=after_sacc,
                          toxname=toxname,
                          expnames=expnames,
                          trials=trials,
                          recalculate=recalculate)
```
