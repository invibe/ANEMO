## example of code

-----------------------------

| Once your data is open you can modify it.
| You will find here two code examples showing you how to do it:

### add new data to the Data

Here, we want to add a new data in the data that we will call `new_data`. It corresponds to `xpr` multiplied by 2, but we want to calculate this new data only for some trials of an experiment

```python
import ANEMO

#------------------------------------------------------------------------------
# open Data
#------------------------------------------------------------------------------
# Dictionary containing all the data
full_data = ANEMO.Data.open(dirpath='dataANEMO', sub='001') 

expnames = ['sub-001'] # list of names of the particular experience
trials = [1,100,110] # list of trials

#------------------------------------------------------------------------------
# add new data
#------------------------------------------------------------------------------
for expname in expnames:
    
    # Dictionary correspond to particular experiment
    data_exp = full_data[expname] 
    
    # Data to particular experiment
    Data = data_exp['Data'] # or data_exp.Data
    
    # add new empty data
    Data['new_data'] = [None]*len(Data)
    
    for t in trials:
        trial_data = Data[Data.trial==t] # trial Data
        Data.loc[(Data.trial==t), 'new_data'] = trial_data['eye1_x_coordinate']*2
    
#------------------------------------------------------------------------------
# save Data
#------------------------------------------------------------------------------
ANEMO.Data.save(dirpath='dataANEMO', data=full_data)
```

### add new events to the Events

Here, we want to add a new event in the data that we will call `Stop_search_MiSacc`. This one corresponds to the event `TargetOn` + 100ms

```python
import ANEMO

#------------------------------------------------------------------------------
# open Data
#------------------------------------------------------------------------------
# Dictionary containing all the data
full_data = ANEMO.Data.open(dirpath='dataANEMO', sub='001') 

#------------------------------------------------------------------------------
# add new events
#------------------------------------------------------------------------------
for expname in full_data.keys():
    
    # Dictionary correspond to particular experiment
    data_exp = full_data[expname] 
    
    # Events to particular experiment
    events = data_exp['Events'] # or data_exp.Events
    
    # add new events
    events['Stop_recherche_MiSacc'] = events['TargetOn']+100

#------------------------------------------------------------------------------
# save Data
#------------------------------------------------------------------------------
ANEMO.Data.save(dirpath='dataANEMO', data=full_data)
```
