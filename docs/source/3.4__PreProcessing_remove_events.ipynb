# remove_events

**Removes the data** between two events

## Test the parameters

-----------------------------

**example of code:** test the function parameters for removing ``the raw position x of the eye in pixels`` between the ``StimulusOff`` and ``TargetOn`` events.

```python
import ANEMO

PreProcessing = ANEMO.PreProcessing(dirpath='dataANEMO', sub='001')

#------------------------------------------------------------------------------
# Parameters:
#------------------------------------------------------------------------------
xname = 'eye1_x_coordinate' # The raw position x of the eye in pixels
trial = 8 # Number of the trial to be transformed
expname = 'sub-001' # Name of the particular experience to be transformed
                    # or list(PreProcessing.data.keys())[0]
Sevents = 'StimulusOff' # Name of the start of the event
Eevents = 'TargetOn' # Name of the end of the event
ref_time='time' # Name of the reference time for Sevents, Events start_event
                # and stop_event
add_stime = 0 # Add time at the start of the event
add_etime = 0 # Add time at the end of the event
start_event = None # Name of the event marking the start of the search
stop_event = None # Name of the event marking the end of the search
toxname = 'xpr__supp_SOff_TOn' # Name of the data to be saved
return_ = True # If True returns the value, else saves it in data

#------------------------------------------------------------------------------
# to apply the function:
#------------------------------------------------------------------------------
xpr__supp_SOff_TOn = PreProcessing.Trial.remove_events(xname=xname,
                                                       trial=trial,
                                                       expname=expname,
                                                       Sevents=Sevents,
                                                       Eevents=Eevents,
                                                       ref_time=ref_time,
                                                       add_stime=add_stime,
                                                       add_etime=add_etime,
                                                       start_event=start_event,
                                                       stop_event=stop_event,
                                                       toxname=toxname,
                                                       return_=return_)

#------------------------------------------------------------------------------
# to retrieve the raw data:
#------------------------------------------------------------------------------
data_exp = PreProcessing.data[expname]
data = data_exp.Data
xpr = data[data.trial==trial][xname].values

#------------------------------------------------------------------------------
# Plot:
#------------------------------------------------------------------------------
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(10, 10/(1.681*2)))

axs[0].plot(xpr, 'k', alpha=.7)
axs[0].set_title('the position x of the eye in pixels: eye1_x_coordinate')
axs[0].set_xlabel('time (ms)')
axs[0].set_ylabel('position of the eye (px)')

axs[1].plot(xpr__supp_SOff_TOn, 'k', alpha=.7)
axs[1].set_title('the position x of the eye in pixels:\n%s'%toxname)
axs[1].set_xlabel('time (ms)')
axs[1].set_ylabel('position of the eye (px)')

plt.tight_layout()
plt.show()
```

## Run it on a data set

----------------------------

**example of code:** run this function on the full data for removing ``the raw position x of the eye in pixels`` between the ``StimulusOff`` and ``TargetOn`` events.

```python
import ANEMO

PreProcessing = ANEMO.PreProcessing(dirpath='dataANEMO', sub='001')

#------------------------------------------------------------------------------
# Parameters:
#------------------------------------------------------------------------------
xname = 'eye1_x_coordinate' # The raw position x of the eye in pixels
Sevents = 'StimulusOff' # Name of the start of the event
Eevents = 'TargetOn' # Name of the end of the event
ref_time='time' # Name of the reference time for Sevents, Events start_event
                # and stop_event
add_stime = 0 # Add time at the start of the event
add_etime = 0 # Add time at the end of the event
start_event = None # Name of the event marking the start of the search
stop_event = None # Name of the event marking the end of the search
toxname = 'xpr__supp_SOff_TOn' # Name of the data to be saved
expnames = 'sub-001' # Name of the particular experience to be transformed
                     # or list(PreProcessing.data.keys())[0]
trials = list(range(1, 16)) # List of the trial to be transformed
recalculate = False # Allows you to indicate if you want to force the
                    # calculation on the trials already processed

#------------------------------------------------------------------------------
# to apply the function:
#------------------------------------------------------------------------------
PreProcessing.Data.remove_events(xname=xname,
                                 Sevents=Sevents,
                                 Eevents=Eevents,
                                 ref_time=ref_time,
                                 add_stime=add_stime,
                                 add_etime=add_etime,
                                 start_event=start_event,
                                 stop_event=stop_event,
                                 toxname=toxname,
                                 expnames=expnames,
                                 trials=trials,
                                 recalculate=recalculate)
```
